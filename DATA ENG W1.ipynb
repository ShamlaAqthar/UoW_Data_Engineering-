### Sets

Sets are unordered collections of unique elements. This means that each element can only appear once in a set, and the elements do not have a defined order. Sets are useful for operations like removing duplicates from a list or performing mathematical set operations like union, intersection, and difference.

You can create a set by using curly braces `{}` or the `set()` constructor.

```python
# Creating a set using curly braces
my_set = {1, 2, 3, 4, 5}
print("My set:", my_set)

# Creating a set from a list with duplicate elements
list_with_duplicates = [1, 2, 2, 3, 4, 4, 5]
unique_elements_set = set(list_with_duplicates)
print("Set from list with duplicates:", unique_elements_set)

# Adding an element to a set
unique_elements_set.add(6)
print("Set after adding an element:", unique_elements_set)

# Trying to add a duplicate element (it won't be added)
unique_elements_set.add(4)
print("Set after trying to add a duplicate:", unique_elements_set)

# Removing an element from a set
unique_elements_set.remove(2)
print("Set after removing an element:", unique_elements_set)

# Set operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union
union_set = set1 | set2
print("Union of set1 and set2:", union_set)

# Intersection
intersection_set = set1 & set2
print("Intersection of set1 and set2:", intersection_set)

# Difference
difference_set = set1 - set2
print("Difference of set1 and set2:", difference_set)
```
